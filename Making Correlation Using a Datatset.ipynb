```python
from scipy.stats import *
import pandas as pd
from IPython.display import display
import seaborn as sns

#Creating  A dataframe
df = pd.DataFrame()

# initializing Columns
f1 = [1,2,3,4,5,6]
f2 = [1,4.8,3,4,5,6]
f3 = [-1,-6.9,-3,-4,-5,-6]
f4 = [2, -5.99,1.612,3.14,13, -20]
f5 = [2,23.66,6,8,10,12]


#inserting data into dataframe
df['A'] = f1
df['B'] = f2
df['C'] = f3
df['D'] = f4
df['E'] = f5
display(df)
print("               Corellation of Dataframe")
display(df.corr())
corrdf = df.corr()
print("                    Heatmap of The Corellation")
hm = sns.heatmap(corrdf)

```
